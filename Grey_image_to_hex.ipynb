```python
import matplotlib.image as mpimg
import numpy as np
import cv2
import binascii
import matplotlib.pyplot as plt  # Import matplotlib.pyplot

image = cv2.imread('grey_scale_resized.jpeg')
image = cv2.resize(image, (100,100))
# image = image_resize(image, height = 100, width = 100)
# gray = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
data = np.uint8(gray / 16)
print(data)
# plt_img = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
plt.imshow(data, cmap='gray', vmin=0, vmax=15)
plt.show()

# Convert each pixel value to its corresponding hex string
hex_data = ["{:01x}".format(pixel) for pixel in data.flatten()]
print(hex_data)
# Club 4 pixel values into 1 hex as 0xAABC
clubbed_hex_data = []
for i in range(0, len(hex_data), 4):
    clubbed_hex = "0x{:01x}{:01x}{:01x}{:01x}".format(
        int(hex_data[i], 16),
        int(hex_data[i + 1], 16),
        int(hex_data[i + 2], 16),
        int(hex_data[i + 3], 16),
    )
    clubbed_hex_data.append(clubbed_hex)

# Print the list of clubbed hex values
print(len(clubbed_hex_data))
print('[%s]' % ', '.join(map(str, clubbed_hex_data)))

# Convert each pixel value to its corresponding hex string
# hex_data = ["{:01x}".format(pixel) for pixel in data.flatten()]

# Print the list of hex values
# print(len(hex_data))
# print(hex_data)
```
